```python
def add_time_cols(df):
#     df['Time'] = pd.DatetimeIndex(df['Created Date']).time
#     df['Time'] = df['Time'].apply(lambda x: str(x)[0:2]+str(x)[3:5])
    
    dt_idx = pd.DatetimeIndex(df['Created Date'])

    df['Week'] = dt_idx.week
    df['Weekday'] = dt_idx.weekday
    df['Month'] = dt_idx.month

    return df
```

```python
def top_n_column(df, sort_column, n):
    tmp = df.groupby(sort_column)[sort_column].count()
    tmp.sort(sort_column,ascending=False)
    topFactors = tmp.head(n)
    
    df = df[df[sort_column].map(lambda x: x in topFactors)]
    
    return df
    
```

```python
apartment = df[df['Location Type'] == 'RESIDENTIAL BUILDING']

apartment = top_n_column(apartment, 'Descriptor', 15)

print(apartment.Descriptor.unique())
```

```python
apartment = add_time_cols(apartment)

apartment.head()

# cold_apt = apartment[apartment['Descriptor'] == 'HEAT']

# ggplot(bad_building, aes('Borough')) + geom_histogram() + theme(axis_text_x=element_text(angle=45, vjust=1, hjust=1))
```
