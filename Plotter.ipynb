```python
from ggplot import *
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
%matplotlib inline

from bokeh import mpl
from bokeh.plotting import output_file, show
```

```python
df = pd.read_csv('reduced_2014.csv', parse_dates=['Created Date'])
```

```python
df1 = df.dropna(subset = ['Descriptor'])
```

```python
df1 = df1[df1['Descriptor'] == 'Loud Music/Party']
# df1 = df1[df1['Descriptor'].str.contains("Loud Music/Party|Noise") == True]
df1 = df1[['Created Date', 'Descriptor', 'Borough', 'Latitude', 'Longitude']]
df1['Week'] = pd.DatetimeIndex(df1['Created Date']).week
df1['ODay'] = pd.DatetimeIndex(df1['Created Date']).dayofyear
df1['Day'] = pd.DatetimeIndex(df1['Created Date']).dayofweek


h = ggplot(aes(x='Week', fill='Day'), data=df1) + geom_bar(binwidth=1, origin=-0.5)
print(h)

```

```python
g = ggplot(aes(x='ODay', fill='Day'), data=df1) + geom_bar(binwidth=7)
print(g)
```
