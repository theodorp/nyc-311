```python
import pandas as pd
import pandaqsl
import datetime
from ggplot import *
%matplotlib inline
```

```python
filename = '311_Service_Requests_from_2014.csv'

# df = pd.read_csv(filename, nrows=100)
# df.info()

df = pd.read_csv(filename, nrows=75000, usecols=['Created Date', 'Complaint Type', \
                'Descriptor', 'Location Type', 'Borough', 'Latitude', 'Longitude'], \
                 parse_dates=['Created Date'])

df['Weekday'] = pd.DatetimeIndex(df['Created Date']).weekday

# df.Descriptor.unique()
df_Man = df[(df['Borough'] == 'MANHATTAN') & (df['Descriptor']=='Loud Music/Party')]
```

```python
ggplot(aes(x='Longitude', y='Latitude', color='Weekday'), data=df_Man) + geom_point()
# ggplot(aes(x='Weekday'), data=df_Man) + geom_density()
```
